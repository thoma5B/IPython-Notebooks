```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os

from scipy.stats.mstats import ttest_ind # return (t-stat, p-val) # mstats library for masked arrays
```

```python
def compute_pvalues(df_protein):
    array_ = []
    for x in df_protein:
        new_row = []
        for row in df_protein[x]:
            try: 
                new_row.append( float(row.replace(',','.')) )
            except: 
                new_row.append( row)
        array_.append(new_row)

    A = np.array(array_)
    #print A
    old_shape = A.shape

    B = A.T.reshape( old_shape[1], 2, -1) 
    PATIENT = B[:,0,:]
    CTRL =  B[:,1,:]
    PATIENT_clean = np.ma.masked_array(PATIENT,np.isnan(PATIENT))
    CTRL_clean = np.ma.masked_array(CTRL,np.isnan(CTRL))
    p_values =  ttest_ind( PATIENT_clean, CTRL_clean , axis = 1)[1]
    return list(p_values), old_shape # , equal_var = False)
```

```python
folder = "/media/thomas/Partition/Mairas_stuff/iPSCs/"
result = {}
for protein in os.listdir(folder):
    df_protein = pd.read_csv(folder + protein, 
                             sep=';', 
                             index_col = 0) #use_cols = ['CTRL',''])
    p_values, old_shape = compute_pvalues(df_protein)
    result[protein[:-4]] = (p_values, 'old_shape:{}'.format(old_shape)) 

```

```python
print '{:>20}{:>10}{:>10}'.format('Protein','p-value', 'sig')
print '{:{fill}>41}'.format(' ', fill = 'm')
for k,v in result.items():
    print '{:>20}'.format(k,)
    for p_value in v[0]:
        if 0.1 < p_value:
            sig = '' 
        elif 0.05 < p_value:
            sig = '(*)' 
        elif 0.03 < p_value:
            sig = '*' 
        elif 0.01 < p_value:
            sig = '**' 
        elif 0.001 < p_value:
            sig = '***' 
        else:
            sig = '****' 
        print '{:>20}{:9.4f}{:>11}'.format('', p_value, sig)
    print '{:{fill}>41}'.format(' ', fill = '_')
```
